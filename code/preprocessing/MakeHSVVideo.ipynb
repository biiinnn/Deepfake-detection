```python
import cv2
import numpy as np
import glob
import os
from tqdm import tqdm

from google.colab import drive
drive.mount('/content/gdrive/')
data_path = '/content/gdrive/Shareddrives/딥러닝_보주콩/data/deepfake-detection-challenge/train_sample_videos/'
original_data_path = '/content/gdrive/Shareddrives/딥러닝_보주콩/data/deepfake-detection-challenge/train_sample_videos/Original/'
hsv_data_path = '/content/gdrive/Shareddrives/딥러닝_보주콩/data/deepfake-detection-challenge/train_sample_videos/HTI/'
```

```python
## original video to hsv transformed video
video_paths = glob.glob(original_data_path + "*.mp4")
print(video_paths)

for vp in tqdm(video_paths):
    video_name = vp.split("/")[-1][:-4]
    vidcap = cv2.VideoCapture(vp)
    success,image = vidcap.read()

    img_array = []
    while success:
        img_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) 
        
        val = 255
        array = np.full(img_hsv.shape, (0,val,0), dtype=np.uint8)
        val_add_image = cv2.add(img_hsv, array)
        val_add_image = cv2.cvtColor(val_add_image, cv2.COLOR_HSV2BGR)
        success,image = vidcap.read()

        img_array.append(val_add_image)
        
        height, width, layer = val_add_image.shape
        size = (width,height)
       
    out = cv2.VideoWriter(hsv_data_path+video_name+'.mp4',cv2.VideoWriter_fourcc(*'DIVX'), 20, size)
        
    for i in range(len(img_array)):
        out.write(img_array[i])
    out.release()

    print("finish! convert video to frame {name}".format(name=video_name))

print("all convert finish!!")
```
